```
def foo(result_img_size):
    ###########PATH#################
    #result_img_size = 60
    folder_path = "drive/MyDrive/Praca_Inzynierska/pismogotyckie/dasrtseldesjdisc00frit_jp2"
    training_path = folder_path + "/TrainingFiles/"+str(result_img_size)+"/"
    writte_path = folder_path + "/TrainingFiles/Final/"+str(result_img_size)+"train_file.csv"
    ###############################

    # module level variables ##########################################################################
    #Size of the digit
    #result_img_size = 60
    #max numbers of one char 
    max_size = 100
    #enable chars
    enable_chars = ['A', 'a', 'B', 'b', 'C', 'c', 'D', 'd', 'E', 'e', 'F', 'f', 'G', 'g', 'H', 'h', 'I', 'i', 'J', 'j', 'K', 'k', 'L', 'l', 'M', 'm', 'N', 'n', 'O', 'o', 'P', 'p',
                    'R', 'r', 'S', 's', 'T', 't', 'U', 'u', 'W', 'w', 'Y', 'y', 'Z', 'z', 'Ä', 'ä', 'Ö', 'ö', 'Q', 'q', 'ẞ', 'ß', 'Ü', 'ü', 'V', 'v', 'X', 'x', '1','2','3', '4', 
                    '5', '6', '7', '8', '9', '0', ',', '.', '(', ')', '-', '\'', '\"','+', '=','#', '^', '*', '?', '[', ']', '%', ':',';',
                    'ch', 'ck', 'st', 'll', 'ss', 'fi', 'si', 'ff', 'fl']
    x = {char: [] for char in enable_chars}
    ###################################################################################################

    _, _, filenames = next(walk(training_path))

    forbidden = 0
    for filename in filenames:
        path = training_path + filename
        data_train = np.loadtxt(path, delimiter=' ', dtype='<U32')
        print("Size of the feature vector", data_train.shape)
        #x = {char: data_train[data_train[:, 0] == char, :]
        #     for char in enable_chars}
    
        for row in data_train:
            try:
                x[row[0]].append(row[:])
            #all_data = np.vstack((all_data, row[:]))

            except:
                forbidden = forbidden + 1
                print(row[0])


    print(forbidden)

    false_added = 40
    from skimage import util
    for element in x:
        if len(x[element]) < max_size and len(x[element]) > 0:
            old_len = len(x[element])
            j = 0
            for i in range(0, false_added-old_len):
                vector = x[element][j][:]
                tmp = vector[1:]      
                tmp = np.asarray(tmp, dtype=np.float_, order='C')
                noise = np.random.normal(0, .01, tmp.shape)
                tmp = tmp + noise
                tmp = np.asarray(tmp, dtype='<U32', order='C')  
                vector[1:] = tmp
                x[element].append(vector)
                j = j + 1
                if j >= old_len:
                    j = 0
        if len(x[element]) > max_size:
            x[element] = x[element][:max_size]

    finall_arr = np.vstack([zz for zz in x.values() if len(zz) > 0])

    f_handle = open(folder_path + "/TrainingFiles/Final/"+str(result_img_size)+"all_train_file.csv",'w')
    np.savetxt(f_handle, finall_arr, delimiter=" ", fmt="%s")
    f_handle.close()
```
